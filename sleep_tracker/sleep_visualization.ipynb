```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the logged sleep data
csv_file = "sleep_log.csv"  
df_sleep = pd.read_csv(csv_file)
```

```python
# Convert Date column to datetime for proper plotting
df_sleep["Date"] = pd.to_datetime(df_sleep["Date"])

# Plot Sleep Duration Over Time
plt.figure(figsize=(12, 6))
plt.plot(df_sleep["Date"], df_sleep["Duration (Hours)"], marker="o", linestyle="-", color="b", label="Sleep Duration")

plt.xlabel("Date")
plt.ylabel("Sleep Duration (Hours)")
plt.title("Sleep Duration Over Time")
plt.xticks(rotation=45)  # Rotate date labels for better visibility
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Plot Sleep Duration as a Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df_sleep["Date"], df_sleep["Duration (Hours)"], color="skyblue")

plt.xlabel("Date")
plt.ylabel("Sleep Duration (Hours)")
plt.title("Daily Sleep Duration")
plt.xticks(rotation=45)

plt.show()
```
